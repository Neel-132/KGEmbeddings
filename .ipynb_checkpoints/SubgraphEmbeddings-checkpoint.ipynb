```python
sgnodesn = rch.drop_duplicate_node(sgnodes)
```

```python
with open('subgraphnew.edgelist', 'w') as s:
    for item in sgnodesn:
        s.write('{} {}\n'.format(item[0], item[1]))
    print('File written successfully')
                 
s.close()
```

```python
edge_f = "subgraphnew.edgelist"
isDirected = True
G = graph_util.loadGraphFromEdgeListTxt(edge_f, directed=isDirected)
G = G.to_directed()
```

```python
embedding = [HOPE(d = 4, beta = 0.01)]
print('Num nodes: %d, num edges: %d' % (G.number_of_nodes(), G.number_of_edges()))
for emb in embedding:
    t1 = time()
    Y = emb.learn_embedding(graph=G, is_weighted=True, no_python=True)
    print(emb.get_method_name()+':\n\tTraining time: %f' % (time() - t1))
    MAP, prec_curv, err, err_baseline = gr.evaluateStaticGraphReconstruction(G, emb, Y, None)
    print(("\tMAP: {} \t precision curve: {}\n\n\n\n"+'-'*100).format(MAP, prec_curv[:5]))
```
