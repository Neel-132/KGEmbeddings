```python
from Gprocess import roach as rch
import matplotlib.pyplot as plt
import tqdm as tqdm
from time import time

from gem.utils import graph_util
from gem.evaluation import visualize_embedding as viz
from gem.evaluation import evaluate_graph_reconstruction as gr

from gem.embedding.hope import HOPE
from gem.embedding.lap import LaplacianEigenmaps
from gem.embedding.lle import LocallyLinearEmbedding
from argparse import ArgumentParser
```

```python
l = []
with open('train.txt') as data:
    for line in data:
        l.append(line)
triple = rch.split_raw_triple(l)
triple
```

```python
triple = rch.delete_edge(triple)

triple
```

```python
key = rch.nodemap(triple)
key
keyinv = rch.mapback(key)
len(key)
```

```python
ntriple = rch.preprocess(triple, key)
ntriple
```

```python
adj = rch.getadjlist(ntriple, keyinv)
adj
```

```python
degreedist = rch.sort_degreedist(rch.getdegreedist(triple), reverse = True)
degreedist
```

```python
sgraph = rch.getsubgraph(triple, [785, 1038],adj, 2)
sgkey = rch.nodemap(sgraph)
```

```python
sgkeyinv = rch.mapback(sgkey)
sgkey
```

```python
sgraph = rch.preprocess(sgraph, sgkey)
```

```python
sgraph = rch.drop_duplicate_node(sgraph)
```

```python
with open('subgraphnew.edgelist', 'w') as s:
    for item in sgraph:
        s.write('{} {}\n'.format(item[0], item[1]))
    print('File written successfully')
                 
s.close()
```

```python
edge_f = "subgraphnew.edgelist"
isDirected = True
G = graph_util.loadGraphFromEdgeListTxt(edge_f, directed=isDirected)
G = G.to_directed()
```

```python
embalg = ['HOPE', 'LaplacianEigenMaps', 'LocallyLinearEmbedding']
with open('op_wordnet.txt', 'w') as op:
    embedding = [HOPE(d = 4, beta = 0.01), LaplacianEigenmaps(d=2), LocallyLinearEmbedding(d=2)]
    print('Num nodes: %d, num edges: %d' % (G.number_of_nodes(), G.number_of_edges()))
    for emb in range(len(embedding)):
        t1 = time()
        Y = embedding[emb].learn_embedding(graph=G, is_weighted=True, no_python=True)
        op.write('\n\n\n{}\n\n\n'.format(embalg[emb]))
        op.write(embedding[emb].get_method_name()+':\n\tTraining time: %f' % (time() - t1))
        MAP, prec_curv, err, err_baseline = gr.evaluateStaticGraphReconstruction(G, embedding[emb], Y, None)
        if embalg[emb] == 'LaplacianEigenMaps':
            op.write(("\tMAP: {} \t precision curve: {}\n\n\n\n"+'-'*100).format(MAP, prec_curv[:50:-1]))
        else:
            op.write(("\tMAP: {} \t precision curve: {}\n\n\n\n"+'-'*100).format(MAP, prec_curv[:200]))
    print('File written successfully')
op.close()
```
